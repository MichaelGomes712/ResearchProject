```python
import scipy.io
import pandas as pd
import numpy as np
import datetime
import cv2
from PIL import Image
import matplotlib.pyplot as plt
from datetime import date,timedelta
import keras,os
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPool2D , Flatten
from keras.preprocessing.image import ImageDataGenerator
```

```python
wikiMat = scipy.io.loadmat('wiki_crop/wiki.mat')
print(wikiMat)
print(wikiMat['wiki']['dob'])
```

```python
wikiDictionary = {'dob' : wikiMat['wiki']['dob'][0][0][0],'photo_taken' : wikiMat['wiki']['photo_taken'][0][0][0],'full_path' : wikiMat['wiki']['full_path'][0][0][0],'gender' : wikiMat['wiki']['gender'][0][0][0],'face_score' : wikiMat['wiki']['face_score'][0][0][0],'second_face_score' : wikiMat['wiki']['second_face_score'][0][0][0]}
wikiDataset = pd.DataFrame(data = wikiDictionary)
ages = []
for i in range(wikiDataset.shape[0]): 
    ages.append(int((date(wikiDataset.photo_taken.iloc[i],7,1)-(date.fromordinal(int(wikiDataset.dob.iloc[i])) + timedelta(days=int(wikiDataset.dob.iloc[i]%1)) - timedelta(days = 366))).days/365))
wikiDataset['age'] = ages;
display(wikiDataset)
```

```python
wikiDataset['second_face_score'] = wikiDataset['second_face_score'].replace(np.nan,'noDel')
wikiDataset = wikiDataset[~wikiDataset.isin([np.nan, np.inf, -np.inf]).any(1)]
wikiDataset = wikiDataset[wikiDataset['age'] >= 0]
wikiDataset = wikiDataset[wikiDataset['age'] <= 100]
display(wikiDataset)
print(wikiDataset.shape)
```

```python
trainWiki = wikiDataset.iloc[:30372,:]
print(trainWiki.shape)
testWiki = wikiDataset.iloc[30373:36882,:]
print(testWiki.shape)
valWiki = wikiDataset.iloc[36883:,:]
print(valWiki.shape)
print(wikiDataset['age'].max()-wikiDataset['age'].min())
print(wikiDataset['age'].max())
print(wikiDataset['age'].min())
```

```python
model = Sequential()
model.add(Conv2D(input_shape=(150,150,3),filters=64,kernel_size=(3,3),padding="same", activation="relu"))
model.add(Conv2D(filters=64,kernel_size=(3,3),padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Flatten())
model.add(Dense(units=4096,activation="relu"))
model.add(Dense(units=4096,activation="relu"))
model.add(Dense(units=101, activation="softmax"))

model.compile(optimizer= "adam", loss=keras.losses.categorical_crossentropy, metrics=['accuracy'])
```
